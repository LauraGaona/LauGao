 ## 1.Serie de Fourier

```python
import matplotlib.pyplot as plt
from scipy import integrate
import math
from sympy import symbols
import matplotlib.pyplot as plt

import sympy as sy
from numpy import linalg as LA
import math as mt
import numpy as np
from sympy import integrate
from sympy.abc import x
from sympy import cos,pi,sin
import matplotlib.pyplot as plt
```

```python
def a0(f,p):
    return 2*(integrate(f(x), (x,0,p)))/p
def ak(f,p,k):
    return 2*integrate(f(x)*cos(2*pi*k*x/p), (x,0,p))/p
def bk(f,p,k):
    return 2*integrate(f(x)*sin(2*pi*k*x/p), (x,0,p))/p
def f(x):
    return 1-x
def g(x):
    return x**2
def FourierSeries(f,p,x0,n):
    resultado = a0(f,p)
    for k in range(1,n):
        resultado = ak(f,p,k)*cos(2*pi*k*x0/p)+bk(f,p,k)*sin(2*pi*k*x0/p)
    return resultado
```

### Funcion $ 1-x$

```python
p = 1
x1 = np.linspace(-2*p,2*p,100)
y1 = np.zeros(len(x1))
n = 6
for i in range(len(x1)):
    y1[i] = FourierSeries(f,p,x1[i],n)
```

$f_2=f(0,0)(-24x_1x_2)|_{x_1=0}x_1 +(-12x_1-6)|_{x_2=0}x_2=-6x_2$
